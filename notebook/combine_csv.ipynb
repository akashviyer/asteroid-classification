```python
import pandas as pd
import numpy as np
```

```python
query = pd.read_csv('data/sbdb_query_results.csv')
```

```python
query.head()
```

```python
col_names = ['pdes', 'name', 'year', 'class', 'score', 'nclass', 'method', 'bad', 'seq', 'moid', 'abs_mag', 'prop_semimaj_ax', 'prop_eccen', 'prop_inclin_sin', 'osc_semimaj_ax', 'osc_eccen', 'osc_inclin']
```

```python
import re
with open('data/sdsstax_ast_table.tab', 'r') as file:
    lines = file.readlines()

processed_lines = []
for line in lines:
    processed_line = line.strip()
    processed_line = re.sub(r'([a-z.])(\s)([a-zA-Z])', r'\1\3', processed_line)
    processed_line = re.sub(r'(-)(\s{11})([A-Z])', r'\1     -     \3', processed_line)
    processed_line = re.split(r'\s+', processed_line)
    processed_lines.append(processed_line)
```

```python
sdss = pd.DataFrame(processed_lines)
```

```python
sdss.head()
```

```python
sdss = sdss.drop(3, axis=1)
```

```python

sdss.columns = col_names

sdss = sdss.replace('-', np.nan)

sdss.head()
```

```python
sdss['pdes'] = sdss['pdes'].astype(int)
```

```python
merged_data = sdss.merge(query, on='pdes', how='inner')
merged_data.head()
```

```python
merged_data.info()
```

```python
merged_data = merged_data.drop(['extent', 'GM', 'IR', 'BV', 'UB', 'spec_B', 'spec_T', 'full_name', 'seq', 'name_x', 'name_y', 'year', 'nclass', 'method', 'moid'], axis=1)
```

```python
merged_data = merged_data.replace('0.0000', np.nan)
merged_data = merged_data.replace('00.0000', np.nan)
```

```python
merged_data = merged_data.rename(columns = {
    'a' : 'semimaj_ax2',
    'e' : 'eccentricity',
    'i' : 'inclin',
    'om': 'asc_node_long',
    'w' : 'peri',
    'ma': 'mean_anomaly'
})
```

```python
merged_data.head()
```

```python
merged_data.to_csv('data/raw_data.csv', index=False, header=True)
```
